```python
import requests
import json

BASE_URL = "http://localhost:8000"

def pretty(response):
    print("Status:", response.status_code)
    try:
        print(json.dumps(response.json(), indent=2))
    except Exception:
        print(response.text)
    print("-" * 50)
```

```python
# Health check
print("Testing root endpoint:")
response = requests.get(f"{BASE_URL}/")
pretty(response)
```

```python
# Get databases
print("Getting list of databases")
response = requests.get(f"{BASE_URL}/databases/")
pretty(response)
```

```python
# Create item
print("Creating item")
payload = {
    "name": "Item Teste",
    "description": "Criado via script de teste"
}
response = requests.post(f"{BASE_URL}/items", json=payload)
pretty(response)

created_item = response.json()
item_id = created_item["id"]


```

```python
# Get all items
print("Getting all items")
response = requests.get(f"{BASE_URL}/items")
pretty(response)
```

```python
#Get single item
print("Getting single item")
response = requests.get(f"{BASE_URL}/items/{item_id}")
pretty(response)
```
